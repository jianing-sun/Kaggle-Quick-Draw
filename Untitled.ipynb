```python
def read_one_class(name='The_Eiffel_Tower', train_split_folder='small_train_set'): 
    class_id = []
    num_class = len(class_array)
    name = name.replace('_', ' ')
    
    # read full df .csv files
    print('\r\t load df   :  %24s'%(name))
    class_df = pd.read_csv(base_path + 'train_%s/%s.csv'%('simplified', name))
    print('\n')
    
    # acquire train split set's id
    print('\r\t load train_set split:   %24s'%(name))
    name = name.replace('_', ' ')
    key_id = np.load(base_path + 'noise_removal/%s/%s.npy'%(train_split_folder, name))
    label = np.full(len(key_id), 0, dtype=np.int64)
    drawing_id = class_df.loc[class_df['key_id'].isin(key_id)].index.values
    class_id.append(np.vstack([label, drawing_id, key_id]).T)

    return class_df, class_id
```

```python
class_df, class_id = read_one_class()
```

```python
small_train_set = TestGenerator(class_df[:20], class_id[0][:20], batch_size=1, norm=normalize, shuffle=False)
```
